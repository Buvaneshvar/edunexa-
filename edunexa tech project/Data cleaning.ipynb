```python
#Cleaning the Entire Dataset Using the applymap Function
university_towns = []
with open("C:\\Users\\divya\\Desktop\\university_towns.txt") as file:
     for line in file:
        if '[edit]' in line:
             # Remember this `state` until the next is found
             state = line
        else:
             # Otherwise, we have a city; keep `state` as last-seen
            university_towns.append((state, line))

university_towns[:5]
```

```python
towns_df = pd.DataFrame(university_towns,
                       columns=['State', 'RegionName'])
towns_df.head()
```

```python
def get_citystate(item):
     if ' (' in item:
         return item[:item.find(' (')]
     elif '[' in item:
         return item[:item.find('[')]
     else:
         return item
towns_df =  towns_df.applymap(get_citystate)
towns_df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_df = sns.load_dataset('titanic')

# Display basic information about the dataset
print(titanic_df.info())

# Display the first few rows of the dataset
print(titanic_df.head())
```

```python
import numpy as np

# Introduce missing values
titanic_df.loc[5:15, 'age'] = np.nan
titanic_df.loc[25:35, 'fare'] = np.nan

# Introduce outliers
titanic_df.loc[50, 'fare'] = 500  # Adding an outlier to 'fare'
titanic_df.loc[100, 'age'] = 200   # Adding an outlier to 'age'
```

```python
# Handle missing values
# Impute missing 'age' values with the mean
titanic_df['age'].fillna(titanic_df['age'].mean(), inplace=True)

# Impute missing 'fare' values with the median
titanic_df['fare'].fillna(titanic_df['fare'].median(), inplace=True)

# Handle outliers
# Clip 'age' values to a reasonable range
titanic_df['age'] = titanic_df['age'].clip(0, 100)

# Clip 'fare' values to a reasonable range
titanic_df['fare'] = titanic_df['fare'].clip(200)

# Display basic information about the cleaned dataset
print(titanic_df.info())

# Display the first few rows of the cleaned dataset
print(titanic_df.head())
```
